```python
import requests

def get_zipcode(latitude, longitude, api_key):
    base_url = "https://api.geoapify.com/v1/geocode/reverse"
    params = {
        "lat": latitude,
        "lon": longitude,
        "apiKey": api_key
    }

    try:
        response = requests.get(base_url, params=params)
        data = response.json()
        
        if "features" in data and len(data["features"]) > 0:
            properties = data["features"][0]["properties"]
            if "postcode" in properties:
                return properties["postcode"]
            elif "postalCode" in properties:
                return properties["postalCode"]

        return None  # ZIP code not found in the response

    except Exception as e:
        print(f"Error: {str(e)}")
        return None

# Example coordinates (replace with your own)
latitude = 70.8224
longitude = -153.9907
api_key = 'eab569d164134e24bd999800c17395d3'

# Get the ZIP code
zipcode = get_zipcode(latitude, longitude, api_key)
if zipcode:
    print(f"ZIP code: {zipcode}")
else:
    print("ZIP code not found.")
```

```python
import csv
import json

# Define the function to convert CSV to JSON
def csv_to_json(csv_file, json_file):
    data = []
    with open(csv_file, mode='r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        for row in csv_reader:
            data.append(row)

    with open(json_file, mode='w') as json_file:
        json.dump(data, json_file, indent=4)

# Specify the file paths (replace with your own)
csv_file_path = '../project-3-brianne/Resources/USGS_polar_bear_maternal_dens_1910-2018.csv'  # Replace with your CSV file path
json_file_path = '../project-3-brianne/Resources/USGS_polar_bear_maternal_dens_1910-2018.json'  # Replace with your desired JSON file path

# Call the function to convert CSV to JSON
csv_to_json(csv_file_path, json_file_path)
```
