```python
# read polarBearSick_2013_2014.csv and make to df
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

polar_bear_data = pd.read_csv('polarBearSick_2013_2014.csv')
polar_bear_data
```

```python
data_2013 = polar_bear_data[polar_bear_data['Year'] == 2013]
data_2014 = polar_bear_data[polar_bear_data['Year'] == 2014]


contaminants = ['sumPCB', 'sumClBz', 'sumHCH', 'sumDDT', 'sumCHL', 'sumOC']


def plot_pie_chart(year_data, year):
    total_bears = len(year_data)
    contaminated_bears = year_data[contaminants].gt(0).any(axis=1).sum()
    healthy_bears = total_bears - contaminated_bears

    labels = ['Contaminated', 'Healthy']
    sizes = [contaminated_bears, healthy_bears]
    colors = ['red', 'blue']

    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
    plt.title(f'Number of Bears in {year}\n(Total: {total_bears})')


plot_pie_chart(data_2013, 2013)


plot_pie_chart(data_2014, 2014)
```
